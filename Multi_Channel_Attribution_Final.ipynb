### Specify the parameters

```
nr_of_samples = 100000
f_opt = [0.5, 1.0]
factor_opt = [0.5, 1.0]
fs = (5, 15)
hyp_func_opt = [np.sinh, np.sin, np.cosh, np.cos, np.tanh, np.tan, np.arctan]
snr_interval = (16, 20)
conversion_rate = 0.4
time_interval = (0, 30) # Number of total days the data was collected

touch_dict = {
    "social": 1,
    "direct": 2,
    "organic": 3,
    "display:": 4,
    "video": 5,
    "email": 6,
    "other": 7
}
```

## Data Creation

### Generate Curves


```
def create_curves(f_list=[0.5, 1.0], factor_list=[0.5, 1.0], fs=(5,10),
                  hyp_funcs=[np.cos, np.tanh], days=(0, 30)): 
    curves = []
    for func in hyp_funcs:
        for f in f_list:
            for factor in factor_list:
                curve = Curve.Curve(func, fs, f, factor, days)
                curves.append(curve)
    return curves

# Create settings for the hyperbolic curves
curves = create_curves(f_opt, factor_opt, fs, hyp_func_opt, time_interval )
for i in range(len(curves)):
    
    curves[i].hyperbolic()
    curves[i].plot_curve()
    print(curves[i].hyp_func)
    curves[i].add_noise(15)
    print(curves[i].time_distribution)
    curves[i].plot_curve()
```

### Generate the touchpoints, times and answers

```
# Create the settings for the touhcpoints
touchpoints = Touchpoints.TouchPoints(curves=curves, touchpoint_names=touch_dict, 
                                      conversion_rate=conversion_rate,
                                      time=time_interval)

# Generate curves with the given settings and the random noise
gen = Generate.Generate(touchpoints, curves, snr_interval=snr_interval)


# Generate distributions form the generated hyperbolic curves 
paths, times, answers = gen.generate_distribution(nr_of_samples)


io_helper = IOhelper.IOhelper("", google_colab=True)
io_helper.write_2d_csv(paths, "paths")
io_helper.write_2d_csv(times, "times")
io_helper.write_1d_csv(answers, "answers")
io_helper.write_dict_to_csv(touchpoints.names, "vocabulary")
io_helper.dump_json(touch_dict, "vocabulary")
```

## Contextual Importance & Utility

### Specify Functions

```
def findCI(cmin, cmax, absmin, absmax):
    """
           cmax - cmin
    CI = ----------------
          absmax - absmin
    """
    return (cmax-cmin)*100/100


def findCU(answer, cmin, cmax):
    """
            y - cmin
    CU = --------------- 
           cmax - cmin
    """
    return (answer-cmin)/(cmax-cmin)
```
